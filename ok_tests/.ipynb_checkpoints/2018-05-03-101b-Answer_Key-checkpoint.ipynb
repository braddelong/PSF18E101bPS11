B.5.1:

(a) Something like:

>(1) flexprice: a low value of $ {\epsilon}_o $: $ 0 = \mu(x_{\epsilon}\Delta{\epsilon}_o) - \mu\left(I_r + x_{\epsilon}{\epsilon}_r\right){\Delta}r $

>* $ {\Delta}r = \frac{x_{\epsilon}}{I_r + x_{\epsilon}{\epsilon}_r}\Delta{\epsilon}_o = \frac{x_{\epsilon}}{I_r + x_{\epsilon}{\epsilon}_r}(-0.1) $ :: lower domestic real interest rates
>* $ {\Delta}I = \frac{{I_r}x_{\epsilon}}{I_r + x_{\epsilon}{\epsilon}_r}(0.1) $ :: higher domestic investment

>(2) flexprice: a low level of $ r^f $: $ 0 = \mu\left(x_{\epsilon}{\epsilon}_r\right)\Delta{r^f} - \mu\left(I_r + x_{\epsilon}{\epsilon}_r\right)r $

>* $ {\Delta}r = \frac{x_{\epsilon}{\epsilon}_r}{I_r + x_{\epsilon}{\epsilon}_r}\Delta{r^f} = \frac{x_{\epsilon}{\epsilon}_r}{I_r + x_{\epsilon}{\epsilon}_r}(-0.02) $ :: lower domestic real interest rates
>* $ {\Delta}I = \frac{{I_r}x_{\epsilon}{\epsilon}_r}{I_r + x_{\epsilon}{\epsilon}_r}(0.02) $ :: higher domestic investment


(b) Something like:

>(1) sticky-price: a low value of $ {\epsilon}_o $: $ {\Delta}Y = \mu(x_{\epsilon}\Delta{\epsilon}_o) $

>* $ {\Delta}Y = \mu(x_{\epsilon})(-0.1)$ :: lower real GDP, and lower exports and consumption spending

>(2) sticky-price: a low level of $ r^f $: $ \Delta{Y} = \mu\left(x_{\epsilon}{\epsilon}_r\right)\Delta{r^f} $

>* $ \Delta{Y} = \mu\left(x_{\epsilon}{\epsilon}_r\right)(-0.02) $ :: lower real GDP, and lower exports and consumption spending

>(3) sticky-price: a high level of $ r $: $ {\Delta}Y = - \mu\left(I_r + x_{\epsilon}{\epsilon}_r\right)r $

>* $ {\Delta}Y = - \mu\left(I_r + x_{\epsilon}{\epsilon}_r\right)r = - \mu\left(I_r + x_{\epsilon}{\epsilon}_r\right)(+0.02) $ :: lower real GDP, lower investment spending, and lower exports and consumption spending

(c) Appropriate words...: Something like this: In the flexible price model, a strong dollar produced by the actions of foreign exchange speculators and foreign central banks produces a domestic economy with a higher level of investment, and presumably a higher rate of economic growth as that investment adds to the capital stock. In the sticky price model, by contrast, a strong dollar—however produced—brings a shortfall of production and high unemployment. Bob Rubin's statement needed to be qualified by an "in the long run, in which wages, prices, and debt have adjusted..." or "as long as the Federal Reserve keeps the interest rate equal to the neutral rate..."
